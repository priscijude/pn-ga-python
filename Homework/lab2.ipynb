```python
sales['Date'] = pd.to_datetime(sales['Date'])
```

```python
sales['Year'] = pd.DatetimeIndex(sales['Date']).year
```

```python
#changing column title 'Sales Rep Number' to 'Sales Rep ID' to have matching Keys

sales.rename(columns={'Sales Rep Number': 'Sales Rep ID'}, inplace=True)
```

```python
#left join - joining sales and sales_rep tables
performance = pd.merge(sales, sales_rep, how='left', on='Sales Rep ID')
```

```python
performance
```

```python
#groupby sales rep number and look at their performance
#change mgmt >> hierarchical instruction if sales rep performance >> manager performance --> if correlated with # of managers
#ewplacing NaN with text like 'None' or '-' 

sales_rep_performance = performance.groupby(['Sales Rep Name']).sum()['Sales']
```

```python
#joining cities into customers

customers_cities = pd.merge(customers,
                            cities[['City Code', 'City', 'Region']],
                            how = 'left',
                            on = 'City Code')


#joining customers_cities into sales
sales_with_cities = pd.merge(sales,
                        customers_cities[['Customer Number', 'Customer', 'City', 'Region']],
                        how = 'left',
                        on = 'Customer Number')

sales_with_items = pd.merge(sales_with_cities,
                        item_master[['Item Number', 'Product Group']],
                        how = 'left',
                        on = 'Item Number')

sales_final = pd.merge(sales_with_items,
                       sales_rep[['Sales Rep ID', 'Manager']],
                        how = 'left',
                        on = 'Sales Rep ID')

# print the head to check for a proper join
sales_final.head(10)
```

```python
# total sales by geographic region
fig, ax = plt.subplots(figsize=(10, 5))
grouped_data = sales_final.groupby('Region').sum().sort_values('Sales', ascending = False)['Sales']
grouped_data.plot(ax = ax, kind = 'bar')
plt.title('Total Sales ($) By Region')

# regional sales over time
#fig, ax = plt.subplots(figsize=(10, 5))
fig, ax = plt. subplots(figsize=(10,5))
annual_regional_sales = sales_final.groupby(['Region','Year']).sum().sort_values('Year', ascending = False)['Sales']
annual_regional_sales .plot(ax = ax, kind = 'bar')
plt.title('Annual Sales ($) By Region')

#plt.title('Annual Sales ($) By Region')
```

```python
# total sales by product group
fig, ax = plt.subplots(figsize=(10, 5))
grouped_data = sales_final.groupby('Product Group').sum().sort_values('Sales', ascending = False)['Sales']
grouped_data.plot(ax = ax, kind = 'bar')
plt.title('Total Sales ($) By Product Group')

# product group sales over time
fig, ax = plt.subplots(figsize=(15, 8))
annual_product_group_sales = sales_final.groupby(['Product Group','Year']).sum().sort_values('Year', ascending = False)['Sales']
annual_product_group_sales .plot(ax = ax, kind = 'bar')
plt.title('Annual Sales ($) By Product Group')
```

# In 2012 to 2014, the top seller product was "Produce" for 3 consecutive years

```python
# total sales by manager
fig, ax = plt.subplots(figsize=(10, 5))
grouped_data = sales_final.groupby('Manager').sum().sort_values('Sales', ascending = False)['Sales']
grouped_data.plot(ax = ax, kind = 'bar')
plt.title('Total Sales ($) By Manager')

# sales per manager over time
fig, ax = plt.subplots(figsize=(15, 8))
annual_manager_sales = sales_final.groupby(['Manager','Year']).sum().sort_values('Year', ascending = False)['Sales']
annual_manager_sales .plot(ax = ax, kind = 'bar')
#plt.title('Annual Sales ($) By Manager')

# sales per manager in the USA
usa_sales = sales_final[sales_final['Region'] == 'USA']
fig, ax = plt.subplots(figsize=(15, 8))
manager_sales_usa = usa_sales.groupby('Manager').sum().sort_values('Sales', ascending = False)['Sales']
manager_sales_usa.plot(ax = ax, kind = 'bar')
plt.title('Annual USA Sales ($) By Manager in the USA')
```
